### Creaci√≥n de histograma usando matriz numpy

```python

import matplotlib.pyplot as plt #importing matplotlib

img = plt.imread('image/flower.png') # lee los datos de la imagen

#creacion de histograma
# calculate frequency of pixels in range 0-255 
plt.hist(img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k') #calculating histogram
```

### Trazado de histogramas utilizando Matplotlib

```python
# importing required libraries of opencv 
import cv2 
from matplotlib import pyplot as plt 

# reads an input image 
img = cv2.imread('image/perro.jpg',0) 

# find frequency of pixels in range 0-255 
histr = cv2.calcHist([img],[0],None,[256],[0,256])

# show the plotting graph of an image 
plt.plot(histr) 
plt.show() 
```

```python
import cv2 
from matplotlib import pyplot as plt 
img = cv2.imread('image/perro.jpg',0) 

# alternative way to find histogram of an image 
plt.hist(img.ravel(),256,[0,256]) 
plt.show() 
```
